# Setup postgres

## Docker command
```docker
docker run -d 
    --name postgres 
    -e POSTGRES_DB=your_dbname 
    -e POSTGRES_USER=your_user 
    -e POSTGRES_PASSWORD=your_password 
    -p 5432:5432 
    postgres

docker exec -it postgres /bin/bash
>>> psql --dbname=your_dbname --username=your_user
```

```sql
your_dbname=#
CREATE TABLE IF NOT EXISTS my_table (
    v1 FLOAT,
    v2 FLOAT
);

your_dbname=#
INSERT INTO my_table (v1, v2)
SELECT random(), random()
FROM generate_series(1, 100);

your_dbname=#
SELECT COUNT(*) FROM my_table;
```

````python
dataset = PostgresDataset(
    db="your_dbname",
    user="your_user",
    password="your_password",
    host="host.docker.internal",
    port="5432",
    query="SELECT v1,v2 FROM my_table;",
    fetch_size=10000,
)
````

````python
import torch

dataloader = PostgresDataLoader(dataset, batch_size=5, dtype=torch.float32)

for batch in dataloader:
    print(batch)
````
